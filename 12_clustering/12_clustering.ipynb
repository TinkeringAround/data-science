# Clustering
Analyse this Dataset with Kmeans

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

plt.rcParams['figure.figsize'] = (16, 9)
```

```python
# Data Import
data = pd.read_csv("./data.csv")

data.head()
```

```python
# Plotting the Dataset
X = data.values

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
```

```python
# Setup KMeans
ks = [2,3,4,5,6,7,8]

for k in ks:
    kmeans = KMeans(n_clusters=k)
    
    # Fitting with inputs
    kmeans = kmeans.fit(X)

    # Predicting the clusters
    labels = kmeans.predict(X)

    # Getting the cluster centers
    C = kmeans.cluster_centers_
    print("Clustering Vector for k=",k, "\n", C, "\n")

    # Plotting
    fig = plt.figure(k)
    ax = Axes3D(fig)
    ax.scatter(X[:, 0], X[:, 1], X[:, 2])
    ax.scatter(C[:, 0], C[:, 1], C[:,2], marker='*', c='#050505', s=1000)
```

# Answer
It seems like the optimal k is between 5 or 6. It tends to overfit when k > 6.
